## Задача 1
<a id='task1'></a>
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ds = pd.read_csv(r'C:\Users\aydem\Desktop\Netology_Homeworks\Netology_Homeworks\Pandas\ml-latest-small\ratings.csv')
ds.head()
   
```

```python
def ratings(row):
    if row['rating'] <=2:
        return 'Low rating'
    elif row['rating'] <=4:
        return 'Average rating'
    else:
        return 'High rating'
```

```python
ds['ratings'] = ds.apply(ratings, axis=1)
```

```python
ds[['userId', 'movieId', 'rating', 'timestamp', 'ratings']]
ds.head(10)
```

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

Результат классификации запишите в отдельный столбец region.

```python
geo = pd.read_csv(r'C:\Users\aydem\Desktop\Netology_Homeworks\Netology_Homeworks\Pandas\ml-latest-small (1)\ml-latest-small\keywords.csv')
geo.head(100)
```

```python
def geotag(row):
    if 'москва' in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'мурманск' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'Undefined'
```

```python
geo['geotags'] = geo.apply(geotag, axis=1)
```

```python
geo[['keyword', 'shows', 'geotags']]
geo.head()
```

```python
geo.geotags.value_counts()
```

```python
years = np.arange(1950,2010)
```

### Сейчас в каждой строчке проходим по всем значениям из years
### Есть ли какой-то менее время затратный способ решения? 

```python
def production_year(row):
    for i in years: 
        if str(i) in row['title']:
            return i    
    else:
        return '1900'
```

```python
movies_names['year'] = movies_names.apply(production_year, axis=1)
```
