## Задача 2

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

```python
#Собственное решение.
#Круто, все работает, но записывает абракадаброй опять
       
'''
- Создаем файл для записи
- проверяем была ли покупка по списку на основе purchase log по purchases из 1 задания
- создаем новую строчку, в которую загружаем категорию по словарю purchases из 1 задания
'''
with open(r'C:\Users\aydem\Desktop\Netology_Homeworks\Netology_Homeworks\005_work_with_files\Downloads\funnel_8.csv','w', encoding = 'utf-8') as result: 
    with open(r'C:\Users\aydem\Desktop\Netology_Homeworks\Netology_Homeworks\005_work_with_files\Downloads\visit_log.csv','r', encoding = 'utf-8') as vis_lg:
        for line in vis_lg:
            user_id =line.strip().split(',')[0]
            if user_id in purchases:
                category = purchases[user_id]
                line_out_2 = line.strip() + ', ' + category + '\r'
                result.write(line_out_2)
```
